```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup

# 용어 사전 url : https://terms.naver.com/ 앞에 추가

result= []
n=0
j
# IT용어사전 - 전체 페이지 용어&정의 추출  
for i in range(200):
  i=i+1   
  html = urlopen("https://terms.naver.com/list.naver?cid=42346&categoryId=42346&page="+str(i))
  bsObject = BeautifulSoup(html, "html.parser")

  f = bsObject.find_all("div", {"class": "info_area"})
  
  for text in f:
    n=n+1
    t = text.find('a').get_text()
    d = text.find('p').get_text()
    result.append([n, t, d])
    print(n, t, d, '\n')

data = pd.DataFrame(result, columns = ['No.', 'terms', 'definition'])
data.to_excel('drive/My Drive/Colab Notebooks/naver01_02_0610.xlsx', index = False)
#data.to_csv('test.csv', encoding='cp949')  
```
